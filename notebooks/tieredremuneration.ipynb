# Tiered Remuneration

This notebook is the clean driver for the tiered-remuneration animation.

The animation logic now lives in `notebooks/tieredremuneration_support.py`, which gives us:
- one place for the numerical construction of `u`, `u'`, and the remuneration schedules
- one place for the repeated visual patterns such as narration and bank-movement markers
- one notebook cell per act, so each section is still easy to inspect and edit
- one export cell that saves both per-act assets and a single concatenated MP4


```python
from pathlib import Path
import sys

from IPython.display import HTML


def find_project_root() -> Path:
    candidates = [Path.cwd(), Path.cwd().parent]
    for candidate in candidates:
        if (candidate / 'src' / 'visualizer').exists() and (candidate / 'notebooks').exists():
            return candidate.resolve()
    raise RuntimeError('Could not find the project root containing src/visualizer and notebooks/.')


PROJECT_ROOT = find_project_root()
NOTEBOOK_DIR = PROJECT_ROOT / 'notebooks'
OUTPUT_DIR = NOTEBOOK_DIR / 'outputs' / 'tieredremuneration'

for extra_path in (PROJECT_ROOT / 'src', NOTEBOOK_DIR):
    extra_path_str = str(extra_path)
    if extra_path_str not in sys.path:
        sys.path.insert(0, extra_path_str)

from tieredremuneration_support import (
    FPS,
    VIDEO_DPI,
    TieredRemunerationBundle,
    build_tiered_remuneration_bundle,
    export_bundle_artifacts,
    render_schedule_html,
)
```

```python
bundle: TieredRemunerationBundle = build_tiered_remuneration_bundle()

for act_name, schedule in bundle.acts.items():
    print(f'{act_name}: {schedule.total_duration:.2f} seconds')
print(f'Combined: {bundle.combined_schedule.total_duration:.2f} seconds')
print(f'Output directory: {OUTPUT_DIR}')
```

```python
def show_act(act_name: str) -> HTML:
    return HTML(render_schedule_html(bundle.acts[act_name], title=act_name, fps=FPS))
```

## Export

This cell saves:
- one HTML file per act
- one MP4 file per act
- one concatenated MP4 for the full talk track

The export uses the reduced plotting grid from the support module, so it is slower than the preview cells but still practical.


```python
artifacts = export_bundle_artifacts(bundle, OUTPUT_DIR, fps=FPS, dpi=VIDEO_DPI)

print('HTML files:')
for act_name, path in artifacts['html'].items():
    print(f'  {act_name}: {path}')

print()
print('Video files:')
for act_name, path in artifacts['video'].items():
    print(f'  {act_name}: {path}')

print()
print(f"Combined video: {artifacts['combined_video']}")
```
